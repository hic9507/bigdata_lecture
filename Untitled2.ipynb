### 제9차_탐색적데이터분석

```python
# Null 데이터 점검
for col in data.columns:
    msg = 'column: {:>10}\t Percent of NaN value: {:.2f}%'.format(col, 100 * (data[col].isnull().sum() / data[col].shape[0]))
    print(msg)
```

```python
import missingno as msno
msno.matrix(df=data.iloc[:, :], figsize=(8,8), color=(0.8, 0.5, 0.2))
```

```python
msno.bar(df=data.iloc[:, :], figsize=(8,8), color=(0.8, 0.5, 0.2))
```

```python
# EDA - 분석 목표(병 여부) 확인
f, ax = plt.subplots(1, 2, figsize=(18, 8))

data['sex'].value_counts().plot.pie(explode=[0, 0.1], autopct='%1.1f%%', ax=ax[0], shadow=True)

ax[0].set_title('Pie plot - output')
ax[0].set_ylabel('')
sns.countplot('output', data=data, ax=ax[1])
ax[1].set_title('Count plot - output')

plt.show()
```

```python
data[['cp', 'output']].groupby(['cp'], as_index=True).count()
```

```python
pd.crosstab(data['cp'], data['output'], margins=True).style.background_gradient(cmap='summer_r')
```

```python
data[['cp', 'output']].groupby(['cp'], as_index=True).mean().sort_values(by='output', ascending=False).plot.bar()
```

```python
f, ax = plt.subplots(1, 2, figsize=(18, 8))
data[['sex', 'output']].groupby(['sex'], as_index=True).mean().plot.bar(ax=ax[0])
ax[0].set_title('output vs sex')
sns.countplot('sex', hue='output', data=data, ax=ax[1])
ax[1].set_title('sex: o or x')
```
