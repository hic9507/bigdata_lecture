```python
import random

ans = random.randrange(1,101,1)
num =-1
t_cnt=0
print("1~100 숫자 업앤다운 게임을 시작합니다~!")
print("----------------------------------")
while(ans!=num):
    num = int(input("1~100 사이 숫자를 입력하세요: "))
    if num > ans:
        print("down")
    elif num < ans:
        print("up")
    t_cnt += 1
print('----------------------------------')
print(t_cnt, "번 만에 답을 맞춤!")
```

```python
import numpy as np
import time

n = 1000
m = 5000

X = np.random.rand(n, m)
W = np.random.rand(n, 1)
Z = np.zeros((1, m))

start_time = time.time()

for i in range(X.shape[1]):
    for j in range(X.shape[0]):
        Z[0][i] += W[j] + X[j][i]
        
end_time = time.time()
print("반복문 이용 실행 시간: ", (end_time - start_time)*1000, "ms")
```

```python
start_time = time.time()
Z = np.dot(W.T, X)
end_time = time.time()
print("numpy 이용 실행시간: ", (end_time - start_time) * 1000, "ms")
```

```python
import matplotlib.pyplot as plt

N = 8
y = np.zeros(N)

x1 = np.linspace(0, 10, N, endpoint=True)
x2 = np.linspace(0, 10, N, endpoint=False)

plt.plot(x1, y, 'o')
plt.plot(x2, y + 0.5, 'o')
plt.ylim([-0.5, 2])
plt.show()
```
